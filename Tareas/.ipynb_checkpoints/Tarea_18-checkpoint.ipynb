## Mapeo Logístico

sistema dinámico llamado mapeo logísitico: $$x_{n+1}=\mu x_{n}(1-x_{n}),\:\:\mu\in[0,4]\:\mathrm{y}\:x\in[0,1]$$
Entonces $f(x_{n})=\mu x_{n}(1-x_{n})$.

Nos interesa el exponente de Lyapunov en función del parámetro $\mu$,además de observar el diagrama de bifurcaciones. recordemos que el exponente de Lyapunov $\lambda$ está dado por: $$\lambda(x_{0})=\lim_{n\to\infty}\frac{1}{n}\sum_{i=0}^{n-1}\log\left|\:f'(x_{i})\right|\:\:\mathrm{con}\:\:x_{i}=f(x_{i-1}).$$

Para obtener el diagrama de bifurcaciones se necesita iterar muchas veces el sistema dinámico pues se grafica $x_{\infty}$ vs $\:x$ donde $x_{\infty}=f^{n}(x)=f\circ f^{n-1}(x)$ para que se pueda aproximar $x_{\infty}$ se puede calcular $x_n$ para $n$ muy grande.

```julia
include("herramientas.jl")
```

```julia
doc"""
Aproxima el exponente de Lyapunov para un sistema dinámico dado por $f(x,t)$, con input (f,$x_0$ condicion inicial,
valores del tiempo que debe ser un arreglo, iteraciones para el coeficiente de Lyapunov)
"""
function Lyap(f,x0,list,n)
    
#Queremos guardar la aproximacion del exponente de Lyap. guardar las iteraciones (para el diagrama de fases) y guardar las bifurcaciones...
    
    L = [] 
    B = [] 
    rB = [] 
    
    for r in list
        x = x0#condicion inicial
        g(z) = f(z,r) #notemos que al dar la r, la funcion ahora solo depende de una variable
        y = herramientas.numDer(g,x,0.1)# Se calcula la derivada de f en x para r fija.
        l = log(abs(y))# valor absoluta de la derivada
        for i in 2:n
            x = f(x,r) # sistema dinámico
            y = herramientas.numDer(g,x,0.1)#calculamos la derivada de f
            l += log(abs(y))#se aplica la definición del exponente de Lyapunov
        end
        push!(L,l/n)#se agrega el nuevo valor del exponente de Lyapunov
        for j in 1:200
            x = f(x,r) #iteramos para usarlo en el diagrama de bifurcaciones
            push!(B,x)
            push!(rB,r)
        end
    end
    return L,rB,B 
end
```

```julia
f(x,r)=r*x*(1-x);  #ahora vamos a ocupar el ya señalado mapeo logístico
```

```julia
r=linspace(1,4,1000)#con valores entre 0 y 4
L=Lyap(f,0.2,r,10000)#para cada r aplicamos lyapunov con la condición inicial x0=0.5 y 10000 iteraciones
```

```julia
using PyPlot
```

```julia
plot(r,L[1],label=L"Lyapunov $\lambda$")
legend();
```

```julia
plot(L[2],L[3],"r.",markersize=0.3)#de manera analoga recordamos que la segunda parte es r y la tercera x_n para n grande
title("Diagrama de bifurcaciones")
xlim(2.4,4);
```
