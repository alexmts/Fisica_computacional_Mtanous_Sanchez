# Método de Bisección

#definimos la función de bisección como se hizo en clase

```julia
function biseccion(a,b)
    p = (a+b)/2
    f(x) = x^3-x^2-x-1
    for i in 1:100
        if  f(a)*f(p) > 0
            a = p
        end
        
        if f(b)*f(p) > 0
            b = p
        end
        p = (a+b)/2
    end
    return p,f(p)
end;
```

```julia
biseccion(3,5)
```

```julia
list9=biseccion(3.0, 4.0);
list10=biseccion(10.0, 20.0);
list11=biseccion(15.0, 30.0);
list12=biseccion(-0.1, 0.1);
```

```julia
list9
```

#estos de arriba son los intervalos, dados para cada condicion inicial

```julia
xlabel("p")
ylabel("f(p)")
title("Convergencia de las raíces por método de la bisección")
grid("on")
plot(list9)
plot(list10)
plot(list11)
plot(list12)

ylim(-100,400)
xlim(-1,1)

show();
```

#Ahora definimos al funcion como se propuso en clase

```julia
function metodo(f,derivada,a)
    x=Sym("x")
    
    list=zeros(Complex64,40);
    x=a;
    for i in 1:40
            @show [x]
        x=x-f(x)/derivada(x);
        list[i]=x;
    end
    
    return x, f(x), derivada(x)
end
```

```julia
metodo(f,derivada,184736)
```

#Sí funciona, pero ahora queremos que el input sea un rango, no un valor inicial

```julia
function  meto2(b,c) 
    x=b:c 
    n=c-b
    out=zeros(n)
    for i in 1:n
        @show  x[i]
    out[i]=metodo(f, derivada, x[i]) 
    end
return out
end
```

```julia
meto2(10,15)
```

#El ultimo inciso es parecido, pero que sólo tome valores de un rango inicial (linspace)...

```julia
function meto3(b,c,d)
   
    lista=meto2(b,c)
   
    t=[]

    for i in 1:length(listb)
        iteraciones=0
        for j in 1:length(t)
            if lista[i]>t[j]+e
                iteraciones=iteraciones+1
            end
            if listb[i]<iteraciones[j]-e
                iteraciones=iteraciones+1
            end
        end
        if iteraciones==length(t)
            push!(t,listb[i])
        end
    end
    return t
end
```

```julia
#Nótese que usamos un vector para definir el intervalo de (b,c)
#Usando Push! quitamos las raices repetidas.
```
