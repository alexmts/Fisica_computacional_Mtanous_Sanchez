__Ejercicio 1:__. Implementa el método de Euler en `Julia`. Debe ser de la forma `metodo_euler(f,x0,t0,tf,h)`. No olvide que no debe de utilizar `SymPy` de ningún modo en este ejercicio. Naturalmente la función `f` es de dos variables. El _output_ de la función debe ser `return listt,listx`. Otra forma quizá más inteligente de implementarlo es definir desde antes `listt`, y que los inputs del método sean de la forma: `metodo_euler(f,x0,listt)`, naturalmente en `listt` va el tiempo inicial $t_0$ y el paso $h$ que utilizará el método.
* Pruebe su nuevo método resolviendo la ecuación:
$$\frac{dx}{dt}=-\alpha x, x(0)=1\text{ y }2, \text{ y } t\in(0,8).$$
con $\alpha>0$, pruebenlo para $\alpha=2.5$, $h=1.0,0.5,0.2,0.01$.
* Gráfique la solución que obtuvo junto con la curva de la solución exacta, asegúrese de que los puntos de la solución aparezcán unidos en la gráfica para obtener una mejor perspectiva. Haga una gráfica para cada condición inicial, que incluya todos los diferentes pasos usados (excepto $h=1$, este gráfiquelo aparte para cada caso), la idea aquí es ver como nos aproximamos a la solución exacta conforme tomamos $h$'s más pequeños.

```julia
using PyPlot
```

```julia
#Primero definimos la funcion "Euler" que recibe g(x,t), condición inicial a, primer tiempo t1, último tiempo t2 y el tamaño de cada subintervalo h
function euler(g, a, t1, t2, h)
    
    #tArray contiene el intervalo de tiempo dividido en round((tf-ti)/h) subintervalos
    #xArray tiene el mismo número de entradas que timeArray, pero con puros ceros
    tArray = linspace(t1, t2, round((t2-t1)/h));
    xArray = zeros(length(tArray));
    
    ##El primer elemento de xArray es la condición inicial
    xArray[1] = a;
   
    #En el for se implementa el método de Euler
    for i in 1:length(tArray)-1
        xArray[i+1] = xArray[i] +h*g(tArray[i], xArray[i]);
        
    end
    return tArray, xArray; 
end;
```

```julia
#Se define la función sobre la que se pide aplicar el método de Euler
function F(t,x)
    a = 2.5;
    return -a*x;
end;
```

```julia
#Se aplica el método de Euler con los cuatro tamaños de subintervalo, para la condición inicial x(0)=1

(tArray1, xArray1) = euler(F, 1.0, 0.0, 8.0, 1.0);
(tArray2, xArray2) = euler(F, 1.0, 0.0, 8.0, 0.5);
(tArray3, xArray3) = euler(F, 0, 0.0, 8.0, 0.2);
(tArray4, xArray4) = euler(F, 1.0, 0.0, 8.0, 0.01);
```

```julia
#Se grafican las tres soluciones numéricas con h!=1.0 y la solución teórica para el paso anterior

xlabel("t");
ylabel("x(t)");
title("Aproximaciones a la solución con método de Euler\nx(0)=1 y h!=1");
grid("on");
plot(tArray2, xArray2, label="h=0.5");
plot(tArray3, xArray3, label="h=0.2");
plot(tArray4, xArray4, label="h=0.01");
plot(tArray4, exp.(-2.5*tArray4), label="Solución exacta");
legend();
```

```julia
#Se grafica la solución numérica con h==1.0 y la solución teórica de la ecuación diferencial
xlabel("t");
ylabel("x(t)");
title("Aproximación a la solución con método de Euler\nx(0)=1 y h=1");
grid("on");
plot(tArray1, xArray1, label="h=1.0");
plot(tArray4, exp.(-2.5*tArray4), label="Solución exacta");
legend();
```

```julia
#Se aplica el método de Euler con los cuatro tamaños de subintervalo, para la condición inicial x(0)=2
(tArray1, xArray1) = euler(F, 2.0, 0.0, 8.0, 1.0);
(tArray2, xArray2) = euler(F, 2.0, 0.0, 8.0, 0.5);
(tArray3, xArray3) = euler(F, 2.0, 0.0, 8.0, 0.2);
(tArray4, xArray4) = euler(F, 2.0, 0.0, 8.0, 0.01);
```

```julia
#Se grafican las tres soluciones numéricas con h!=1.0 y la solución teórica para el paso anterior
xlabel("t");
ylabel("x(t)");
title("Aproximaciones a la solución con método de Euler\nx(0)=2 y h!=1");
grid("on");
plot(tArray2, xArray2, label="h=0.6");
plot(tArray3, xArray3, label="h=0.3");
plot(tArray4, xArray4, label="h=0.01");
plot(tArray4, 2*exp.(-2.5*tArray4), label="Solución exacta");
legend();
```

```julia
#Se grafica la solución numérica con h==1.0 y la solución teórica de la ecuación diferencial
xlabel("t");
ylabel("x(t)");
title("Aproximaciones a la solución con método de Euler\nx(0)=2 y h=1");
grid("on");
plot(tArray1, xArray1, label="h=1.0");
plot(tArray4, 2*exp.(-2.5*tArray4), label="Solución exacta");
legend();
```

__Ejercicio 2:__. ¿Qué sistema físico está modelado por $\ddot{x} =
-kx$?
* Escriba la solución exacta en terminos de las condiciones iniciales $x(0)$ y $x'(0)$.


## Solución

Suponiendo que $ k>0$ y haciendo $ \omega^{2}=k$, tenemos un oscilador armónico. Sabemos que su solución es:
$$ x(t) = Acos(\omega t) + Bsen(\omega t)$$
$$ \dot{x}(t) = -A\omega sen(\omega t) + B\omega(\omega t)$$

ahora usando las condiciones iniciales obtenemos:

$$ x(0) = A$$
y
$$ \dot{x}(0) = B\omega$$

Así, tenemos que la solucion general es:

$$ x(t) = x(0)cos(\omega t) + \frac{\dot{x}(0)}{\omega}sen(\omega t)$$

El oscilador armónico modela el movimiento de una masa en un resorte de constante $k$ o un péndulo simple con una amplitud inicial pequeña. También sirve para aproximar potenciales en regiones pequeñas.

__Ejercicio 3:__. Haz $u = \dot{x}(t)$ y calcula la derivada de
$u$ ¿qué obtienes y como se relaciona con el sistema

Sea $u = \dot{x}(t)$ entonces se tiene que $\dot{u}=\ddot{x}(t)=-kx$. Se tienen entonces las dos ecuaciones:

$$\left\{
\begin{array}{ll}
      \dot{x}(t)=u \\
      \dot{u}=-kx \\
\end{array} 
\right.$$

__Ejercicio 4:__. Obten la solución del sistema $\ddot{x} = -kx(t)$
utilizando lo escrito arriba y el método de Euler.


```julia
#Método de Euler para varias variables dado por el profesor
function euler(f,list,x0)
     x = x0
     h = list[2]-list[1]
     listx=[]
     push!(listx,x)
     for i in 2:length(list)
        t = i*h
        x = x + f(x,t)*h
        push!(listx,x) 
     end
     return listx
end
```

```julia
#Se define la función para resolver con k=1
f(x,t)=[x[2], -1.0*x[1]]
```

```julia
#Se graficará sobre el intervalo [0,8]
list=0:0.01:8;
#En vec se guardan las soluciones y se dan como condiciones iniciales x(0)=0 y u(0)=2
vec=euler(f,list,[0,2]);
```

```julia
#Se guardan las soluciones en dos lists y la solución teórica en listxreal
listx1=map(x->x[1],vec);
listx2=map(x->x[2],vec);
listxreal=2*sin.(list); 
listxdotreal=2*cos.(list);
```

```julia
#Se grafican las aproximaciones y las soluciones teóricas
xlabel("t")
ylabel("x y u")
title("Aproximaciones a la derivada con la derivada numérica")
grid("on")
plot(list,listx1, label="x");
plot(list,listx2, label="u");
plot(list,listxreal, label="Solución teórica");
plot(list,listxdotreal, label="Derivada de solución teórica");
legend();
```

__Ejercicio 5:__ Aplicar el método de Euler para resolver la ecuación:
$$\frac{dx}{dt}=\frac{2x-18 t}{1+t}.$$
Compare su respuesta con la solución exacta: $4+8t-5t^2$

```julia
#Se define la función sobre la que se pide aplicar el método de Euler
function F2(t,x)
    return (2x-18t)/(1+t);
end;
```

```julia
#Se define la función que soluciona el problema
function fSol(t)
    return 4+8*t-5*(t*t);
end;
```

```julia
#Se aplica el método de Euler con los cuatro tamaños de subintervalo, para la condición inicial x(0)=4
(tArray1, xArray1) = euler(F2, 4.0, 0.0, 8.0, 1.0);
(tArray2, xArray2) = euler(F2, 4.0, 0.0, 8.0, 0.5);
(tArray3, xArray3) = euler(F2, 4.0, 0.0, 8.0, 0.2);
(tArray4, xArray4) = euler(F2, 4.0, 0.0, 8.0, 0.01);
```

```julia
#Se grafican las cuatro soluciones numéricas y la solución teórica para el paso anterior
xlabel("t");
ylabel("x(t)");
title("Aproximaciones a la solución con método de Euler\nx(0)=4");
grid("on");
plot(tArray1, xArray1, label="h=1.0");
plot(tArray2, xArray2, label="h=0.5");
plot(tArray3, xArray3, label="h=0.2");
plot(tArray4, xArray4, label="h=0.01");
plot(tArray4, fSol.(tArray4), label="Solución exacta");
legend();
```
