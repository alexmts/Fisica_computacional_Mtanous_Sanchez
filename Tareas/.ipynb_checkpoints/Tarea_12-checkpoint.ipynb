__Ejercicio 1:__ Implementar método implicito de Euler como quieran.

```julia
using PyPlot
```

```julia
#Se define una función sobre la que se aplicará el método de Euler
#Esta no depende de "t"
function f(x,t)
    a = 2.5;
    return -a*x;
end;
```

```julia
#encontramos la derivada parcial respecto a "x" de la función f
function df(x,t)
    a = 2.5;
    return -a;
end;
```

```julia
#Se define otra función sobre la que se quiere aplicar el método de Euler
#Esta SI depende de "x" y de "t"
function g(x,t)
    return (2x-18t)/(1+t);
end;
```

```julia
#Se calcula la derivada parcial respecto a "x" de la función g
function dg(x,t)
    return (2)/(1+t);
end;
```

```julia
#Se define la función solución del problema con g(x,t)
function S(t)
    return 4+8*t-5*(t*t);
end;
```

```julia
#Con base en f y df se define la función a la que se le van a calcular las raíces en el método de Euler implicito
function F(y,y0,h,t)
    return y-y0-h*f(y,t);
end;
```

```julia
#Se calcula la derivada parcial respecto a "y" de la función F
function dF(y,y0,h,t)
    return 1-h*df(y,t);
end;
```

```julia
#Análogamente, se define la función a la que le van a calcular las raíces en el método de Euler implicito
function G(y,y0,h,t)
    return y-y0-h*g(y,t);
end;
```

```julia
#Se calcula la derivada parcial respecto a "y" de la función G
function dG(y,y0,h,t)
    return 1-h*dg(y,t);
end;
```

```julia
#Actualizamos el Método de Newton, para que funcione con las funciones F y G
#F es la función en la que se aplica el método, dF su derivada, "y" es la suposición de raíz, "t" es el tiempo para la función de dos variables
function Root(F, dF, y, t, y0, h)
    x=y
       for i in 1:2000
       x = x-(F(x, y0, h, t)/dF(x, y0, h, t))
       end
       return x;
end;
```

```julia
#### Método implícito de Euler ####
#Recibe la función F=y[n+1]-y[n]-h*f(y[n],t[n]), la derivada dF, tiempo inicial t0, tiempo final tn, condición inicial y(t0)=y0 y el tamaño de cada subintervalo h
function impEuler(F, dF, t0, tn, y0, h)
    #Se encuentra el número de subintervalos y se le asigna a n
    n = trunc(Int,(tn - t0)/h);
    
    #Se crean dos array, uno para t y otro para y en los que se va a guardar la solución.
    timeArray = zeros(n+1);
    yArray = zeros(n+1);
    
    #La primera entrada de los arrays es la condición inicial
    timeArray[1] = t0
    yArray[1] = y0
    
    #En el for se encuentra y[n+1] con el método de Newton anterior
    #Además, se llena el array de tiempo
    
    for i in 1:n
        timeArray[i+1]=timeArray[i] + h;
        yArray[i+1] = Root(F,dF,1,timeArray[i+1], yArray[i], h);
    end;
    
    return timeArray, yArray;
    
    ##Necesitamos estos arrays para graficar...
end;
```

```julia
#Se aplica el método implícito en F con las condiciones que se piden... i.e. h=0.1 t y(0)=1
(timeArrayf, xArrayf) = impEuler(F, dF, 0, 10, 1, 0.1);
```

```julia
#Se grafican las 2 soluciones para poder compararlas... la solución numérica con h=0.1 y la solución teórica para F
xlabel("t");
ylabel("x(t)");
title("Aproximación a la solución con método de Euler implicito\nx(0)=1 y h=0.1");
grid("on");
plot(timeArrayf, xArrayf, label="h=0.1");
timeArray = linspace(0,10,2000);
plot(timeArray, exp.(-2.5*timeArray), label="Solución exacta");
legend();
```

```julia
#Se aplica el método implícito en G con h=0.1 t y(0)=4
(timeArrayg, xArrayg) = impEuler(G, dG, 0, 10, 4, 0.1);
```

```julia
#Se grafica la solución numérica con h=0.1 y la solución teórica para G
xlabel("t");
ylabel("x(t)");
title("Aproximación a la solución con método de Euler implicito\nx(0)=1 y h=0.1");
grid("on");
plot(timeArrayg, xArrayg, label="h=0.1");
timeArray = linspace(0,10,2000);
plot(timeArray, S.(timeArray), label="Solución exacta");
legend();
```

__Ejercicio 2:__ Implementar Euler: Regla del punto medio.

```julia
### Ahora usaremos la Regla del punto medio ###
#La función recibe g(x,t), condición inicial a, primer tiempo t1, último tiempo tf y el tamaño de cada subintervalo h
function pm(g, a, ti, tf, h)
    #timeArray contiene el intervalo de tiempo dividido en round((tf-ti)/h) subintervalos
    #xArray tiene el mismo número de entradas que timeArray, pero con puros ceros
    timeArray = linspace(ti, tf, round((tf-ti)/(h)));
    xArray = zeros(length(timeArray));
    #El primer elemento de xArray es la condición inicial
    xArray[1] = a;
    #En el for se implementa la regla del punto medio, que llena las entradas de xArray con t[n+1/2]=t[n]+h/2
    for i in 1:length(timeArray)-1
        xArray[i+1] = xArray[i] +h*g(xArray[i]+(h/2)*g(xArray[i],timeArray[i]), timeArray[i]+(h/2));
    end
    #Se regresan los dos arrays
    return timeArray, xArray; 
end;
```

```julia
#Se aplica la regla del punto medio con cuatro tamaños de subintervalo, para la condición inicial x(0)=1 en f
#Es el test para la regla del punto medio
(timeArray1, xArray1) = pm(f, 1.0, 0.0, 8.0, 0.75);
(timeArray2, xArray2) = pm(f, 1.0, 0.0, 8.0, 0.5);
(timeArray3, xArray3) = pm(f, 1.0, 0.0, 8.0, 0.2);
(timeArray4, xArray4) = pm(f, 1.0, 0.0, 8.0, 0.01);
```

```julia
#Se grafican las cuatro soluciones numéricas y la solución teórica para el paso anterior
xlabel("t");
ylabel("x(t)");
title("Aproximaciones a la solución con método de Euler\nx(0)=1");
grid("on");
plot(timeArray1, xArray1, label="h=0.75");
plot(timeArray2, xArray2, label="h=0.5");
plot(timeArray3, xArray3, label="h=0.2");
plot(timeArray4, xArray4, label="h=0.01");
plot(timeArray4, exp.(-2.5*timeArray4), label="Solución exacta");
legend();
```

__Ejercicio 3:__ Implementar método de Runge-Kutta de orden 4.

```julia
#Runge-Kutta de orden 4
#La función recibe g(x,t), condición inicial a, primer tiempo t1, último tiempo tf y el tamaño de cada subintervalo h
function RK4(g, a, ti, tf, h)
    #timeArray contiene el intervalo de tiempo dividido en round((tf-ti)/h) subintervalos
    #xArray tiene el mismo número de entradas que timeArray, pero con puros ceros
    timeArray = linspace(ti, tf, round((tf-ti)/(h)));
    xArray = zeros(length(timeArray));
    #El primer elemento de xArray es la condición inicial
    xArray[1] = a;
    #En el for se calculan los ks de Runge-Kutta y 
    #se llenan las entradas de xArray con x[n+1]=x[n]+(h/6)(k1+2k2+2k3+k4)
    for i in 1:length(timeArray)-1
        k1 = g(xArray[i], timeArray[i]);
        k2 = g(xArray[i]+(h/2)*k1, timeArray[i]+(h/2));
        k3 = g(xArray[i]+(h/2)*k2, timeArray[i]+(h/2));
        k4 = g(xArray[i]+h*k3, timeArray[i+1]);
        xArray[i+1] = xArray[i]+(h/6)*(k1+2*k2+2*k3+k4);
    end
    #Se regresan los dos arrays
    return timeArray, xArray; 
end;
```

```julia
#Método de Euler explícito
#La función recibe g(x,t), condición inicial a, primer tiempo t1, último tiempo tf y el tamaño de cada subintervalo h
function expEuler(g, a, ti, tf, h)
    #timeArray contiene el intervalo de tiempo dividido en round((tf-ti)/h) subintervalos
    #xArray tiene el mismo número de entradas que timeArray, pero con puros ceros
    timeArray = linspace(ti, tf, round((tf-ti)/h));
    xArray = zeros(length(timeArray));
    #El primer elemento de xArray es la condición inicial
    xArray[1] = a;
    #En el for se implementa el método de Euler, que llena las entradas de xArray
    for i in 1:length(timeArray)-1
        xArray[i+1] = xArray[i] +h*g(xArray[i], timeArray[i]);
    end
    #Se regresan los dos arrays
    return timeArray, xArray; 
end;
```

```julia
#Se aplica el método de Euler con los cuatro tamaños de subintervalo... usando la condición x(0)=1
(timeArray1, xArray1) = expEuler(f, 1.0, 0.0, 8.0, 1.0);
(timeArray2, xArray2) = expEuler(f, 1.0, 0.0, 8.0, 0.5);
(timeArray3, xArray3) = expEuler(f, 1.0, 0.0, 8.0, 0.2);
(timeArray4, xArray4) = expEuler(f, 1.0, 0.0, 8.0, 0.01);
```

__Ejercicio 6__: Grafique la energía total del oscilador harmonico usando los métodos explícito de Euler y Runge-Kutta de orden 4. Compare sus respuestas y discuta.

__Nota:__ Deberán hacer una implementación de Runge-Kutta para dos dimensiones (o de una vez más si quieren, basandose en el ejemplo que les di, si se les ocurre una mánera mejor, adelante).

__Nota:__ Usen $h=0.1$ para hacer la comparación o cualquier otro que refleje una diferencia significativa entre los dos métodos en la energía, el chiste es no usar un $h$ muy pequeño.

```julia
#Método explícit0 de Euler para varias variables... Lo dio el profesor en clase...
function euler(f,list,x0)
     x = x0
     h = list[2]-list[1]
     listx=[]
     push!(listx,x)
     for i in 2:length(list)
        t = i*h
        x = x + f(x,t)*h
        push!(listx,x) 
     end
     return listx
end
```

```julia
#Se define la función para resolver con k=1 y m=1, que surge de resolver d2x/dt2=x
l(x,t)=[x[2], -1.0*x[1]]
```

```julia
#Se graficará sobre el intervalo [0,8] con h = 0.1
list=0:0.1:8;
#En vec se guardan las soluciones con Euler y se dan como condiciones iniciales x(0)=0 y u(0)=2
vec=euler(l,list,[0,2]);
```

```julia
#Se guardan las soluciones en dos lists y la solución teórica en listxreal
listx1euler=map(x->x[1],vec);
listx2euler=map(x->x[2],vec);
listxreal=2*sin.(list); 
listxdotreal=2*cos.(list);
```

```julia
#Se grafican las aproximaciones y las soluciones teóricas
xlabel("t")
ylabel("x y u")
title("Aproximaciones a la solución con método de Euler")
grid("on")
plot(list,listx1euler, label="x");
plot(list,listx2euler, label="u");
plot(list,listxreal, label="Solución teórica");
plot(list,listxdotreal, label="Derivada de solución teórica");
legend();
```

```julia
#Método de Runge-Kutta de orden 4 para varias variables... Utilizando como BASE lo que dio el Profesor
function RK(f,list,x0)
     x = x0
     h = list[2]-list[1]
     listx=[]
     push!(listx,x)
     for i in 2:length(list)
        t = i*h
        k1 = f(x,t);
        k2 = f(x+(h/2)*k1,t+(h/2));
        k3 = f(x+(h/2)*k2, t+(h/2));
        k4 = f(x+h*k3, t+h);
        x = x+(h/6)*(k1+2*k2+2*k3+k4);
        push!(listx,x) 
     end
     return listx
end
```

```julia
#Se graficará sobre el intervalo [0,8] con h = 0.1
list=0:0.1:8;
#En vec se guardan las soluciones con Runge-Kutta y se dan como condiciones iniciales x(0)=0 y u(0)=2
vec=RK(l,list,[0,2]);
```

```julia
#Se guardan las soluciones en dos lists y la solución teórica en listxreal
listx1rk=map(x->x[1],vec);
listx2rk=map(x->x[2],vec);
listxreal=2*sin.(list); 
listxdotreal=2*cos.(list);
```

```julia
#Se grafican las aproximaciones y las soluciones teóricas
xlabel("t")
ylabel("x y u")
title("Aproximaciones a la solución con Runge-Kutta")
grid("on")
plot(list,listx1rk, label="x");
plot(list,listx2rk, label="u");
plot(list,listxreal, label="Solución teórica");
plot(list,listxdotreal, label="Derivada de solución teórica");
legend();
```

```julia
#La energía teórica es constante y es E=0.5*x^2+0.5xdot^2, ya que m=k=1. Encontramos entonces que E=2
function cnt(x,t)
    return 2;
end;
```

```julia
#En esta función se evaluarán las soluciones numéricas, siguiendo que E=0.5*x^2+0.5xdot^2
function energia(x,y)
    return 0.5*x*x+0.5y*y;
end;
```

```julia
#Se grafican las aproximaciones y la solución teórica para la energía del oscilador armónico
xlabel("t")
ylabel("E")
title("Aproximaciones a la energía con Euler y Runge-Kutta")
grid("on")
plot(list,energia.(listx1rk, listx2rk), label="Runge-Kutta");
plot(list,energia.(listx1euler, listx2euler), label="Euler");
plot(list,cnt.(list,list), label="Solución teórica");
legend();
```

SI hay una diferencia entre la solución teórica y la obtenida con Euler. 
De manera numérica, se puede aproximar una recta con una pendiente de $0.25$ (cuando buscamos que la pendiente sea cero), ya que la solución teórica es una constante. Sin embargo, se puede decir que $0.25\approx0$, pero JUSTO aquí notamos el error que tiene el método de Euler. 

Por otra parte, en la gráfica no se observa la aproximación con Runge-Kutta... a continuación se grafica unicamente este metodo

```julia
#Se grafican la aproximación con Runge-Kutta y la solución teórica
xlabel("t")
ylabel("E")
title("Aproximación a la energía con Runge-Kutta")
grid("on")
plot(list,energia.(listx1rk, listx2rk), label="Runge-Kutta");
plot(list,cnt.(list,list), label="Solución teórica");
legend();
```

Observamos aquí que la pendiente de la recta con Runge-Kutta se aproxima en magnitud a $1x10^{-5}\approx0$, que es cuatro órdenes de magnitud menor que con el método de Euler!!! 
Por tanto, a mayor orden del método de Runge-Kutta, mejor será la aproximación, como era de esperarse y que este es un método más preciso que el de Euler explícito.
